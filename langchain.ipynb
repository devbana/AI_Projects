```python
# Imports
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Pick a free Hugging Face chat model
# Gemma-2B is lightweight and free to use
model_name = "tiiuae/falcon-7b-instruct"
```

```python
# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Create a text-generation pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=256,
    temperature=0.7,
    do_sample=True
)

```

```python
# Wrap pipeline with LangChain
llm = HuggingFacePipeline(pipeline=pipe)

# Example: single-turn chat
prompt = "You are a helpful tutor. Explain quantum entanglement simply."
response = llm.invoke(prompt)
print(response)
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
system_template = "Translate the following into {language}:"
```

```python
prompt_template = ChatPromptTemplate.from_messages(
    [("system", system_template), ("user", "{text}")]
)
```

```python
prompt = prompt_template.invoke({"language": "italian", "text": "hi"})
```

```python
retriever = faiss_index.as_retriever(k=3)
```

```python

template = """Answer the question based only on the following context:
{context}

Question: {question}
"""
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
from langchain.schema.runnable import RunnablePassthrough
```

```python
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
```

```python
chain.invoke("what is RAG?")
```

RAG refers to Retriever-Augmented Generation models, which use text documents as additional context when generating target sequences.
